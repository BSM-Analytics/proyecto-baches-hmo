# Imports 

```python
import requests
import os
import zipfile
import pandas as pd
import shutil
```

# Códigos para extraer los datos de nivel socioeconómico

Analizamos datos de varios años, pero finalmente decidimos trabajar con los del 2020.

Pruebas para descargar y extraer datos de nivel socioeconómico 2010

```python
# URL del archivo .zip
url = 'https://www.inegi.org.mx/contenidos/programas/ccpv/2010/datosabiertos/ageb_y_manzana/resageburb_26_2010_csv.zip'
# Nombre del archivo .zip local
zip_filename = 'resageburb_26_2010_csv.zip'
# Directorio donde se extraerán los archivos
extract_dir = os.path.dirname(os.getcwd()) + '/data/raw'

# Descargar el archivo .zip
response = requests.get(url)
with open(zip_filename, 'wb') as f:
    f.write(response.content)

# Crear el directorio de extracción si no existe
if not os.path.exists(extract_dir):
    os.makedirs(extract_dir)

# Extraer el archivo .zip
with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f"Archivo {zip_filename} extraído en {extract_dir}")

# Borrar el archivo .zip después de extraer
os.remove(zip_filename)
```

Pruebas para descargar y extraer datos de nivel socioeconómico 2020

```python
directorio = os.path.dirname(os.getcwd())
url = 'https://www.inegi.org.mx/contenidos/programas/ccpv/2020/datosabiertos/ageb_manzana/ageb_mza_urbana_26_cpv2020_csv.zip'
datos = '/data/raw'
archivo_ageb = '/data/raw/socioeconomico_2020.zip'

if not os.path.exists(directorio + '/data/raw/socioeconomico_2020.zip'):
    if not os.path.exists(directorio + '/data/raw'):
        if not os.path.exists(directorio + '/data'):
            os.makedirs(directorio + '/data')
        os.makedirs(directorio + '/data/raw')
    r = requests.get(url, allow_redirects=True) 
    open(directorio + '/data/raw/socioeconomico_2020.zip', 'wb').write(r.content)

    with zipfile.ZipFile(directorio + archivo_ageb) as zip_ref:
            zip_ref.extractall(directorio + datos)

if not os.path.exists(directorio + datos + '/conjunto_de_datos_ageb_urbana_26_cpv2020.csv'):
    shutil.move(directorio + datos + '/ageb_mza_urbana_26_cpv2020/conjunto_de_datos/conjunto_de_datos_ageb_urbana_26_cpv2020.csv'
                , directorio+datos)
if not os.path.exists(directorio + datos + '/diccionario_datos_ageb_urbana_26_cpv2020.csv'):
    shutil.move(directorio + datos + '/ageb_mza_urbana_26_cpv2020/diccionario_de_datos/diccionario_datos_ageb_urbana_26_cpv2020.csv'
                , directorio + datos)

shutil.rmtree(directorio + datos + '/ageb_mza_urbana_26_cpv2020')
os.remove(directorio + archivo_ageb)
```
